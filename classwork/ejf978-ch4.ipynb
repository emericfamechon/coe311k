```python
import numpy as np
import math
import random
import matplotlib.pyplot as plt
```

```python
pi = np.pi
leibseries = 0
for i in range(0, 5000):
    leibresult = ((-1)**i)/(2*i+1)
    pileib = 4*leibresult
    leibseries = leibseries + pileib
    relerror = np.abs(np.pi - leibseries)
    if i%100 == 0:
        print("Estimate", leibseries)
        print("Error", relerror)
```
