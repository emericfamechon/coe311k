```python
import numpy as np
import math
import random
import matplotlib.pyplot as plt
```

montecarlo function (from earlier challenges done)

```python
def montecarlo_pi(number_of_shots):
    number_of_hits = 0
    for i in range(0,number_of_shots):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if ((x*x + y*y) < 1):
            number_of_hits = number_of_hits+1
    return 4 * (number_of_hits/number_of_shots)
```

```python
truerror = []
truepi = []
steps = []

for i in range (10, 1000):
    y = montecarlo_pi(i)
    truepi = np.pi
    truerror = np.append(truerror, abs(y-truepi))
    steps = np.append(steps, i)
    i = i+20
```

```python
plt.plot(steps, truerror)
```
