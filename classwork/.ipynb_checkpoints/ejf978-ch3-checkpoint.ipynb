```python
import numpy as np
import math
import random
import matplotlib.pyplot as plt
```

```python
def montecarlo_pi(number_of_shots):
    number_of_hits = 0
    for i in range(0,number_of_shots):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if ((x*x + y*y) < 1):
            number_of_hits = number_of_hits+1
    return 4 * (number_of_hits/number_of_shots)
```

```python
truerror = []
relerror = []
truepi = []

for i in range(0,100):
    y = montecarlo_pi(1000)
    truepi = np.pi
    truerror = np.append(truerror, abs(truepi - y))
    relerror = np.append(relerror, abs((truepi - y)/truepi))
    print(truerror, relerror)
```

```python
plt.subplot(2, 2, 1) 
plt.hist(truerror, 100)
plt.show()

plt.subplot(2,2,2)
plt.hist(relerror, 100)
plt.show()
```
