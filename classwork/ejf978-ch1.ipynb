```python
import numpy as np
import matplotlib.pyplot as plt

z = np.linspace(0, 2*np.pi, 50)
y = np.random.rand(1000)
x = np.random.rand(1000)
h = np.random.rand(1000)

plt.subplot(2, 2, 1) 
plt.plot(z, np.cos(z), 'r-+')
plt.plot(z, np.tan(z), 'k-')

plt.subplot(2, 2, 2)
y = np.tan(z)
plt.scatter(z,y)

plt.subplot(2, 2, 3)
plt.hist(y, 50)

plt.subplot(2, 2, 4)
size = np.random.rand(1000) * 50
color = np.random.rand(1000)
plt.scatter(x, h, size, color)
plt.colorbar()

plt.show()
```
