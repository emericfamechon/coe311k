```python
num = 2025
twobit = ""

def find_bits(num, twobit):
    finalbit = ""
    while (num) > 0:
        modulo = num % 2
        if (modulo == 1):
            twobit = str(1)+twobit
        elif (modulo == 0):
            twobit = str(0)+twobit
        num = num//2
        finalbit = twobit
    return finalbit

result = find_bits(num,twobit)

print(result)
            
```
